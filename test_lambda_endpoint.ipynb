## Amazon API Gateway経由でテスト

```python
# Amazon API Gateway 経由の場合、そのままバイナリデータを渡す
data = f.read()
```

```python
header = {
    'x-api-key':'XXXXX',
    'Content-Type':'image/jpeg'
}
```

```python
response = requests.post(
    "https://XXXXX.execute-api.ap-northeast-1.amazonaws.com/default/vehicle-endpoint-test", 
    data=data, 
    headers=header
)
```

```python
imageBody =  base64.b64decode(response.content)
Image.open(io.BytesIO(imageBody))
```
